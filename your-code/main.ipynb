### 1. Function to Generate Uniform Random Numbers
Using SciPy (as your exercise suggests):

```python
def generate_uniform(bottom, ceiling, count):
    # loc = bottom, scale = ceiling - bottom
    return uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)
```

### 2. Generate the Two Required Datasets

```python
# First dataset
data1 = generate_uniform(10, 15, 100)

# Second dataset
data2 = generate_uniform(10, 60, 1000)
```

```python
data2
```

```python
data1
```

### 3. Plot Histograms (10 bins each)

```python
plt.figure(figsize=(12, 5))

# Plot 1
plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, edgecolor='black')
plt.title("Uniform Distribution: 10 to 15 (100 samples)")
plt.xlabel("Value")
plt.ylabel("Count")

# Plot 2
plt.subplot(1, 2, 2)
plt.hist(data2, bins=10, edgecolor='black')
plt.title("Uniform Distribution: 10 to 60 (1000 samples)")
plt.xlabel("Value")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

How are the two distributions different?
### Key Difference Between the Two Uniform Distributions
Even though both are uniform, they do not look the same because:

#### 1. Different sample sizes
* First distribution: **100 samples**
* Second distribution: **1,000 samples**

With only 100 points, randomness creates **visible bumps and dips**.
With 1,000 points, the histogram becomes **much smoother** and closer to a perfect rectangle.

This is the Law of Large Numbers in action.

#### 2. Different ranges
* First: **10 → 15** (range = 5)
* Second: **10 → 60** (range = 50)

A wider range spreads the values out more, so:
* Bars are **lower** because the same number of samples is spread across a larger interval.
* The distribution looks **flatter** and more stretched.

### Summary
Even though both are uniform distributions:

* The **small sample** looks noisy.
* The **large sample** looks smooth.
* The **wide range** spreads values out more.

The underlying distribution is the same — the **appearance** changes because of **sample size** and **interval width**.

```python
from scipy.stats import norm 
```

### 1. Function to Generate Normal Random Numbers

```python
def generate_normal(mean, std, count):
    return norm.rvs(loc=mean, scale=std, size=count)
```

### 2. Generate the Two Required Datasets

```python
# Dataset 1: mean=10, std=1
normal_1 = generate_normal(mean=10, std=1, count=1000)

# Dataset 2: mean=10, std=50
normal_2 = generate_normal(mean=10, std=50, count=1000)
```

### 3. Plot the Distributions (Histograms)

```python
plt.figure(figsize=(12, 5))

# Plot 1
plt.subplot(1, 2, 1)
plt.hist(normal_1, bins=30, edgecolor='black')
plt.title("Normal Distribution: mean=10, std=1")
plt.xlabel("Value")
plt.ylabel("Count")

# Plot 2
plt.subplot(1, 2, 2)
plt.hist(normal_2, bins=30, edgecolor='black')
plt.title("Normal Distribution: mean=10, std=50")
plt.xlabel("Value")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

### How are the two distributions different?

* **The mean is the same**, so both are centered around 10
* **The standard deviation is very different**

  * `std = 1` → narrow, tall, tightly clustered around 10
  * `std = 50` → extremely wide, flat, spread across a huge range

This matches the expected output image.

### How to tell if a variable is “nearly normally distributed”

A variable is approximately normal when:
* Its histogram looks **bell‑shaped**
* It is **symmetric** around the mean
* It has **one clear peak** (unimodal)
* It does not have long tails or extreme skew
* It does not have many outliers

We can also confirm with:
* A Q–Q plot
* A normality test (Shapiro, Anderson-Darling, KS)

## **Exponential Distribution**

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

### 1. Function to Generate Exponential Samples (mean = 10)

```python
def generate_exponential(mean, count):
    # numpy uses scale = mean for exponential distribution
    return np.random.exponential(scale=mean, size=count)
```

### 2. Generate Two Sequences (size 10 and 100)

```python
data_10 = generate_exponential(mean=10, count=10)
data_100 = generate_exponential(mean=10, count=100)
```

### 3. Plot Histograms (bins = 100)

```python
plt.figure(figsize=(14, 5))

# Plot 1 — 10 samples
plt.subplot(1, 2, 1)
plt.hist(data_10, bins=100, edgecolor='black')
plt.title("Exponential Distribution (mean=10, size=10)")
plt.xlabel("Value")
plt.ylabel("Count")

# Plot 2 — 100 samples
plt.subplot(1, 2, 2)
plt.hist(data_100, bins=100, edgecolor='black')
plt.title("Exponential Distribution (mean=10, size=100)")
plt.xlabel("Value")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

### Key Differences Between the Two Exponential Distributions
#### 1. Sample size
* **10 samples** → extremely noisy, irregular, jagged
* **100 samples** → smoother, clearer exponential decay

Small samples don’t capture the true shape of the distribution.
Larger samples begin to reveal the expected pattern.

#### 2. Shape clarity
* With **10 points**, the histogram may look almost random
* With **100 points**, you start to see:
    
    * A high bar near 0
    * Rapid decline toward the right
    * The classic exponential “drop‑off”

This is the Law of Large Numbers in action.

#### 3. Stability
* The **10‑sample** distribution changes drastically every time you run it
* The **100‑sample** distribution is more stable and consistent

More data → less randomness → more reliable shape.

### Summary
The distribution with 10 samples is very irregular because the sample size is too small to show the true exponential pattern. The distribution with 100 samples is much smoother and clearly shows the exponential decay shape. Larger samples always produce a more accurate representation of the underlying distribution.

The mean changes, so the distribution changes as well. 

## **Exponential Distribution of Real Data**

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

### Problem

Time spent in a bank is **exponentially distributed** with:
* Mean = 10 minutes
* Rate ( \lambda = 1/10 = 0.1 )

We want:

$$ P(X < 15) $$

For an exponential distribution:

$$ P(X < x) = 1 - e^{-\lambda x} $$

### Python Code (using NumPy)

```python
# parameters
mean = 10
lam = 1 / mean

# probability that X < 15
x = 15
prob = 1 - np.exp(-lam * x)

prob
```

### Interpretation

This computes:

$$ P(X < 15) = 1 - e^{-0.1 \cdot 15} $$

Which evaluates to approximately:

$$ 1 - e^{-1.5} \approx 0.7769 $$

So a customer has about a **77.7% chance** of spending less than 15 minutes in the bank.

### What is the probability that the customer will spend more than 15 minutes
The probability that a customer spends **more than 15 minutes** in the bank is simply the complement of spending **less than 15 minutes**.

Since we already computed:

$$ P(X < 15) = 1 - e^{-0.1 \cdot 15} $$

Then:

$$ P(X > 15) = 1 - P(X < 15) $$

```python
import numpy as np

mean = 10
lam = 1 / mean

x = 15

# Probability customer spends MORE than 15 minutes
prob_more = np.exp(-lam * x)

prob_more
```

### Interpretation

This computes:

$$ P(X > 15) = e^{-1.5} \approx 0.2231 $$

So there is about a **22.3% chance** that a customer spends **more than 15 minutes** in the bank.
